```python
import requests
from pprint import pprint
import pandas as pd
import time as t

def none_decode(x, y):
    if x == None:
        return y
    else:
        return x


url = "https://api.zp.ru/v1/vacancies"
lim_value = 100
df = pd.DataFrame(columns=['company', 'position', 'location', 'format', 'salary', 'salary_min',
       'salary_max', 'gross_ind', 'schedule', 'levels', 'techstack',
       'date_publish', 'id', 'requirements','requirements_short'])
#url = 'https://api.zarplata.ru/vacancies'

# этой функцией industries можно получить справочник для параметра industry, 
# в частности id: '7' это 'Информационные технологии, системная интеграция, интернет'


for i in range(50):
    params = {'scope':'public',
             'limit':lim_value,
             'fields':'header,company.title',
             'industry':7,
             'date_from':'2020-12-01',
             'date_to':'2020-12-02',
             'describe_arguments':'true',
              'per_page':100,
             'offset':i*lim_value
             }

    r_cycle = requests.request('GET', url, params = params)
    rj_cycle = r_cycle.json() # Применяем метод json()
    #print(r_cycle.status_code)
    vac_list = list()
    if r_cycle.status_code != 200:
        df.to_csv('file_name.csv', index=False, sep = '\t', encoding='utf-8', na_rep = 'не определено')
        print(f'bad response code {r_cycle.status_code}')
        break
    else: 
        pprint(rj_cycle)
        for vac in rj_cycle['vacancies']:

            location = lambda: 'не определено' if vac['address'] == None else vac['address']['city']['title']
            work_type = lambda: 'не определено' if vac['working_type'] == None else vac['working_type']['title']
            schedule = lambda: 'не определено' if vac['schedule'] == None else vac['schedule']['title']
            #key_skills = ','.join([vac['key_skills'].values()])
            df_row = {'company': vac['company']['title'],
                      'position': vac['header'],
                      'location': location(),
                      'format': schedule(), 
                      'salary': round((none_decode(vac['salary_max'],0) + none_decode(vac['salary_min'],0))/2),
                      'salary_min': vac['salary_min'],
                      'salary_max': vac['salary_max'],
                      'gross_ind': vac['is_salary_gross'],
                      'schedule': work_type(),
                      'levels': vac['experience_length'],
                      'techstack': vac['requirements_short'],
                      'date_publish': vac['add_date'],
                      'id': vac['id'],
                      'requirements': vac['requirements'],
                      'requirements_short':vac['requirements_short']
                     }
            vac_list.append(df_row)

        df_cur = pd.DataFrame(vac_list)
        df = pd.concat([df, df_cur], ignore_index=True, axis=0 )
        print(f'Iteration {i}: dataframe size {df.shape}, status_code is {r_cycle.status_code}')
        t.sleep(6)
    
#df.to_csv('file_name.csv', index=False, sep = '\t', encoding='utf-8', na_rep = 'не определено')
```

3.2) Ранее мы не делали ораничение по ключевым словам, собирали все вакансии, теперь сделам фильтарцию по ключевым полям.
Видим, что вкансий снуными нам скиллами совсем немного, не будем добавлять их в общий датасет, так как ручная проверка показывает, что 
они вряд ли относятся именно в DS/ML-сегменту рынка труда.

```python
import re
def ml_search(x):
    if x == None:
        x=''
    search_lst = ['Data Science','Data Analyst','Data Engineer','Machine Learning','MLOps','Аналитик данных','анализ данных','Инженер по данным',
                 'машинное обучение','нейронные сети', 'моделиров','математ', 'Аналитик по данным', 'big data','большие данные','большим данным',
                 'больших данных']
    #columns_to_search = ['position','techstack_y','requirements']
    res = '-99'
    for search_str in search_lst:
        if re.search(search_str.lower(), x.lower()) != None:
            res = search_str.lower()
        #if 
    return res

df_all['ml_position'] = df_all['position'].apply(ml_search)
df_all['ml_techstack_y'] = df_all['techstack_y'].apply(ml_search)
df_all['ml_requirements'] = df_all['requirements'].apply(ml_search)

df_all[(df_all['ml_position'] != '-99') | (df_all['ml_techstack_y'] != '-99') | (df_all['ml_requirements'] != '-99')][['id','position']]
```
